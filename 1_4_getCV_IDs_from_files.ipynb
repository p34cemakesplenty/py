```python
#############################################################
# RETRIEVES UNIQUE CV ID FROM                               #
# EXISTING CV FILES AND COUNT                               #
# @PARAM: UN/EDITED CV FILES                                #
# @RETURN: C:/Users/jozes/Python_Files/_CV_data/MyCvIDs.txt #
#############################################################
import re
import glob
import os.path

CV_Set=set()
cv_count=0
print('hello')
tot_dupe_count=0
for filename in glob.glob('_CV_data/CV_*'):
    f_dupe_count=0
    print (filename)
    print('Running count: '+str(cv_count))
    with open(filename, "rt",encoding='utf8') as fin:
        contents=fin.read()
        f_entries_count=0
        #print(type(contents))
        #print(contents.find_all('CV_'))
        for m in re.finditer('CV_',contents):
            #print('CV_ found', m.start(), m.end())
            cv=contents[m.start()+3:m.start()+19]
            f_entries_count+=1
            #print('CV_:'+cv)
            if cv in CV_Set:
                 print('dupe: '+cv)
            else:
                CV_Set.add(cv)
                cv_count+=1
        for n in re.finditer('\[',contents):
            #print('[ found so', n.start(), n.end())
            if contents[n.start()-17]==']':
                cv=contents[n.start()-16:n.start()-0]
                f_entries_count+=1
                #print('CV:'+cv)
                if cv in CV_Set:
                    #print('dupe: '+cv+' in file '+filename)
                    f_dupe_count+=1
                else:
                    CV_Set.add(cv)
                    cv_count+=1
    print('Number of dupes in file' +filename+': '+str(f_dupe_count))
    print('Number of entries in file' +filename+': '+str(f_entries_count))
    tot_dupe_count=tot_dupe_count+f_dupe_count
print(len(CV_Set))
with open("C:/Users/jozes/Python_Files/_CV_data/MyCvIDs.txt","w",encoding="utf-8") as f:
    f.write('')
with open("C:/Users/jozes/Python_Files/_CV_data/MyCvIDs.txt","a",encoding="utf-8") as f:
    for cv_id in CV_Set:
        #print(cv_id)
        f.write(str(cv_id)+'\n')
print('finished writing the file C:/Users/jozes/Python_Files/_CV_data/MyCvIDs.txt')
#print(CV_Set)
#print(len(CV_Set))
print('Number of CVs: '+str(cv_count))
print('Total number of dupes in all files: '+str(tot_dupe_count))
```
